```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn;
seaborn.set()
```

```python
plt.figure(figsize=(16,9))
goog.plot();
```

```python
fig = plt.figure(figsize=(16,8))
goog.plot(alpha=0.9, style='-')
goog.resample('BA').mean().plot(style=':')   #BA is B year end
goog.asfreq('BA').plot(style='--');
plt.legend(['input','resample','asfreq'],
          loc='upper left');
```

```python
fig = plt.figure(figsize=(16,8))
goog.plot(alpha=0.5, style='-')
goog.resample('BQS').mean().plot(style=':')   #BA is B year end
goog.asfreq('BQS').plot(style='--');
plt.legend(['input','resample','asfreq'],
          loc='upper left');
```

```python
fig = plt.figure(figsize=(16,8))

fig, ax = plt.subplots(2, sharex=True)
data = goog.iloc[:10]

#missing data
data.asfreq('D').plot(ax=ax[0], marker='*')

data.asfreq('D',method='bfill').plot(ax=ax[1], style='-o')
data.asfreq('D',method='ffill').plot(ax=ax[1], style='--*')
ax[1].legend(["back-fill", 'forward-fill']);
```

```python
fig = plt.figure(figsize=(16,8))

fig, ax = plt.subplots(2, sharex=True)
data = goog.iloc[:120]

#missing data
data.asfreq('D').plot(ax=ax[0], marker='*')

data.asfreq('D',method='bfill').plot(ax=ax[1], style='-o')
data.asfreq('D',method='ffill').plot(ax=ax[1], style='--*')
ax[1].legend(["back-fill", 'forward-fill']);
```

# Time-shifts

```python
plt.figure(figsize=(16,9))
fig, ax = plt.subplots(3, sharey=True)

# apply a frequency to the data
goog = goog.asfreq('D', method='pad')

goog.plot(ax=ax[0])
goog.shift(900).plot(ax=ax[1])
goog.tshift(900).plot(ax=ax[2])

# Legends and annotations
local_max = pd.to_datetime('2007-12-05')
offset = pd.Timedelta(900,'D')

ax[0].legend(['input'], loc=2)
ax[0].get_xticklabels()[2].set(weight='heavy', color='red')
ax[0].axvline(local_max, alpha=0.9, color='red')

ax[1].legend(['shift(900)'], loc=2)
ax[1].get_xticklabels()[2].set(weight='heavy', color='red')
ax[1].axvline(local_max + offset, alpha=0.3, color='red')

ax[2].legend(['tshift(900)'], loc=2)
ax[2].get_xticklabels()[2].set(weight='heavy', color='red')
ax[2].axvline(local_max + offset, alpha=0.3, color='red');
```

# ROI(return of investment): Net profit/total investment*100

```python
fig = plt.figure(figsize=(16,8))
ROI = 100 * (goog.shift(365)/ goog-1)
ROI.plot()
plt.ylabel('% Return on investment');
```

```python
fig = plt.figure(figsize=(16,8))
ROI = 100* (goog.tshift(365)/ goog-1)
ROI.plot()
plt.ylabel('% Return on investment');
```

```python
# Rolling Window
```

```python
plt.figure(figsize=(16,8))
rolling = goog.rolling(365,centre=False)

data = pd.DataFrame({'input': goog,
                    'one-year rolling_mean': rolling.mean(),
                    'one-year rolling_std': rolling.std()
                    })
ax = data.plot(styles=['-', '--', ':'])
ax.lines[0].set_alpha(0.5)
```

```python
by_time = data.groupby(data.index.time).sum()
by_time.head()
```

```python
plt.figure(figsize=(16,9))
hourly_ticks = 3 * 60*60 *np.arange(8)
by_time.plot(xticks=hourly_ticks, style=[':', '--', '-']);
```

```python
by_weekday = data.groupby(data.index.dayofweek).sum()
by_weekday.head()
```

```python
by_weekday.index = ['Mon', 'Tues', 'Wed', 'Thurs', 'Fri', 'Sat', 'Sun']
by_weekday.plot(style=[':', '--', '-']);
```

```python
weekend = np.where(data.index.weekday < 5, 'Weekday', 'Weekend')
weekend[0:50]
```

```python
by_time = data.groupby([weekend, data.index.time]).mean()
by_time.head()
```
